```python
import numpy as np
import matplotlib.pyplot as plt



#read input file
from numpy import loadtxt
file = open('./plot_data/chaufrey_h2_vel.cvs', 'rb')
data = loadtxt(file,delimiter = ",")
file = open('./plot_data/chaufrey_h3p_vel.cvs', 'rb')
data2 = loadtxt(file,delimiter = ",")
file = open('./plot_data/chaufrey_h3p_int.cvs', 'rb')
data3 = loadtxt(file,delimiter = ",")
file = open('./plot_data/chaufrey_rot_vel.cvs', 'rb')
data4 = loadtxt(file,delimiter = ",")
file = open('./plot_data/chaufrey_zero_vel.cvs', 'rb')
data5 = loadtxt(file,delimiter = ",")

zero=np.mean(data5[:,1])


n_rows=31
h3_vel=np.zeros((2,n_rows))
h3_vel_max=np.zeros((2,n_rows))
h3_vel_min=np.zeros((2,n_rows))             

h3_int=np.zeros((2,n_rows))
rot=np.zeros((2,n_rows))

for i in range(n_rows):
    rot[0,i]=i+19-33
    rot[1,i]=zero-data4[i,1]
    
    h3_vel[0,i]=data2[i*3,2]+19-33
    h3_vel[1,i]=zero-data2[i*3,1]-rot[1,i]
    h3_vel_max[0,i]=data2[i*3+1,2]+19-33
    h3_vel_max[1,i]=zero-data2[i*3+1,1]-rot[1,i]
    h3_vel_min[0,i]=data2[i*3+2,2]+19-33
    h3_vel_min[1,i]=zero-data2[i*3+2,1]-rot[1,i] 

    h3_int[0,i]=i+19-33
    h3_int[1,i]=(zero-data3[i,1])*(-1)


n_rows=29
# print(data)

h2_vel=np.zeros((2,n_rows))
h2_vel_max=np.zeros((2,n_rows))
h2_vel_min=np.zeros((2,n_rows))

for i in range(n_rows):
    h2_vel[0,i]=data[i*3,2]+19-32
    h2_vel[1,i]=zero-data[i*3,1]-rot[1,i]
    h2_vel_max[0,i]=data[i*3+1,2]+19-32
    h2_vel_max[1,i]=zero-data[i*3+1,1]-rot[1,i]
    h2_vel_min[1,i]=zero-data[i*3+2,1]-rot[1,i]  


    

    
plt.figure()
plt.plot(h2_vel[0,:],h2_vel[1,:],color='red')
ax1 = plt.axes()
ax1.set_xlabel('Row position (0.353" pixels)')
ax1.set_ylabel('V$_{PRF}$ (km/s)')
ax1.set_ylim(-6, 6)
plt.fill(np.append(h2_vel[0,:],np.flipud(h2_vel[0,:])),np.append(h2_vel_max[1,:],np.flipud(h2_vel_min[1,:])),c='orange',alpha=0.1)

# plt.plot(h2_vel_max[0,:],h2_vel_max[1,:],color='red',linestyle='dashed')
# plt.plot(h2_vel_min[0,:],h2_vel_min[1,:],color='red',linestyle='dashed')

plt.plot(h3_vel[0,:],h3_vel[1,:],color='blue')

plt.fill(np.append(h3_vel[0,:],np.flipud(h3_vel[0,:])),np.append(h3_vel_max[1,:],np.flipud(h3_vel_min[1,:])),c='lightblue',alpha=0.1)

# plt.plot(h3_vel_max[0,:],h3_vel_max[1,:],color='blue',linestyle='dashed')
# plt.plot(h3_vel_min[0,:],h3_vel_min[1,:],color='blue',linestyle='dashed')

plt.plot(rot[0,:],rot[1,:]-rot[1,:],color='black',linestyle='dashed')
plt.plot(h3_int[0,:],h3_int[1,:]*1.8+4,color='black',linestyle='dotted')
plt.savefig('outplots/chaufray_vel1.pdf', facecolor='white', edgecolor='none')
print(rot[0,:])

plt.figure()
ax1 = plt.axes()
ax1.set_xlabel('Row position (0.353" pixels)')
ax1.set_ylabel('V$_{PRF}$ (km/s)')
ax1.set_ylim(-6, 6)
plt.plot(h3_vel[0,0:28],h3_int[1,0:28]*1.6+3,color='black',linestyle='dotted')
# plt.plot(h2_vel[0,1:34],h2_vel[1,1:34],color='red')
plt.plot(h3_vel[0,0:28],h3_vel[1,0:28]-h2_vel[1,0:28],color='green')
plt.plot(rot[0,0:28],rot[1,0:28]-rot[1,0:28],color='black',linestyle='dashed')
plt.savefig('outplots/chaufray_cur1.pdf', facecolor='white', edgecolor='none')
```

```python
import numpy as np
import matplotlib.pyplot as plt



#read input file
from numpy import loadtxt
file = open('./chaufrey_h2_vel2.cvs', 'rb')
data = loadtxt(file,delimiter = ",")
file = open('./chaufrey_h3p_vel2.cvs', 'rb')
data2 = loadtxt(file,delimiter = ",")
file = open('./chaufrey_h3p_int2.cvs', 'rb')
data3 = loadtxt(file,delimiter = ",")
file = open('./chaufrey_rot_vel2.cvs', 'rb')
data4 = loadtxt(file,delimiter = ",")
file = open('./chaufrey_zero_vel2.cvs', 'rb')
data5 = loadtxt(file,delimiter = ",")

zero=np.mean(data5[:,1])


n_rows=35
rot=np.zeros((2,n_rows))

for i in range(n_rows):
    rot[0,i]=i+16-32
    rot[1,i]=zero-data4[i,1]


n_rows=34
h3_vel=np.zeros((2,n_rows))
h3_vel_max=np.zeros((2,n_rows))
h3_vel_min=np.zeros((2,n_rows))             

h3_int=np.zeros((2,n_rows))



for i in range(n_rows):
    
    h3_vel[0,i]=data2[i*3,2]+17-32
    h3_vel[1,i]=zero-data2[i*3,1]-rot[1,i+1]
    h3_vel_max[0,i]=data2[i*3+1,2]+17-32
    h3_vel_max[1,i]=zero-data2[i*3+1,1]-rot[1,i+1]
    h3_vel_min[0,i]=data2[i*3+2,2]+17-32
    h3_vel_min[1,i]=zero-data2[i*3+2,1]-rot[1,i+1] 

    h3_int[0,i]=i+17-32
    h3_int[1,i]=(zero-data3[i,1])*(-1)

n_rows=33
# print(data)

h2_vel=np.zeros((2,n_rows))
h2_vel_max=np.zeros((2,n_rows))
h2_vel_min=np.zeros((2,n_rows))

for i in range(n_rows):
    h2_vel[0,i]=data[i*3,2]+16-32
    h2_vel[1,i]=zero-data[i*3,1]-rot[1,i]
    h2_vel_max[0,i]=data[i*3+1,2]+16-32
    h2_vel_max[1,i]=zero-data[i*3+1,1]-rot[1,i]
    h2_vel_min[0,i]=data[i*3+2,2]+16-32
    h2_vel_min[1,i]=zero-data[i*3+2,1]-rot[1,i]  


    
plt.figure(figsize=(8,6))
ax1 = plt.axes()
ax1.set_ylim(-6, 6)
ax1.set_xlabel('Spatial Axis \nRow Position (0.353$^{\prime\prime}$ per pixel)', fontsize=14)
ax1.set_ylabel('V$_{\mathrm{PRF}}$ (km/s)', fontsize=14)

plt.plot(h2_vel[0,:],h2_vel[1,:],color='red' , label='H$_2$')
# plt.plot(h2_vel[0], h2_vel_max[1], c='y')
# plt.plot(h2_vel[0], h2_vel_min[1], c='g')
plt.fill(np.append(h2_vel[0,:],np.flipud(h2_vel[0,:])),np.append(h2_vel_max[1,:],np.flipud(h2_vel_min[1,:])),color='red', alpha=0.1)

# plt.plot(h2_vel_max[0,:],h2_vel_max[1,:],color='red',linestyle='dashed')
# plt.plot(h2_vel_min[0,:],h2_vel_min[1,:],color='red',linestyle='dashed')

plt.plot(h3_vel[0,:],h3_vel[1,:],color='blue', label='H$_3^{+}$')
# plt.plot(h3_vel[0], h3_vel_max[1], c='y')
# plt.plot(h3_vel[0], h3_vel_min[1], c='g')
plt.fill(np.append(h3_vel[0,:],np.flipud(h3_vel[0,:])),np.append(h3_vel_max[1,:],np.flipud(h3_vel_min[1,:])),color='blue', alpha=0.1)

# plt.plot(h3_vel_max[0,:],h3_vel_max[1,:],color='blue',linestyle='dashed')
# plt.plot(h3_vel_min[0,:],h3_vel_min[1,:],color='blue',linestyle='dashed')

plt.plot(rot[0,:],rot[1,:]-rot[1,:],color='black',linestyle='dashed')
plt.plot(h3_int[0,:],h3_int[1,:]*1.6+3,color='black',linestyle='dotted',lw=2)
plt.title('CML = 179', fontsize=14)
plt.legend(loc=2)

# plt.savefig('outplots/chaufray_vel2.pdf', facecolor='white', edgecolor='none')
# plt.savefig('chaufray_vel2.png', dpi=400, bbox_inches='tight', facecolor='white')

print(rot[0,:])

plt.figure(figsize=(8,6))
ax1 = plt.axes()
ax1.set_ylim(-6, 6)
ax1.set_xlabel('Spatial Axis \nRow Position (0.353$^{\prime\prime}$ per pixel)', fontsize=14)
ax1.set_ylabel('V$_{\mathrm{PRF}}$ (km/s)', fontsize=14)
plt.title('CML = 179', fontsize=14)
plt.plot(h3_vel[0,0:32],h3_int[1,0:32]*1.6+3,color='black',linestyle='dotted',lw=2)
# plt.plot(h2_vel[0,1:34],h2_vel[1,1:34],color='red')
plt.plot(h3_vel[0,0:32],h3_vel[1,0:32]-h2_vel[1,1:33],color='green', label='H$_3^{+}$-H$_2$')
plt.plot(rot[0,1:33],rot[1,1:33]-rot[1,1:33],color='black',linestyle='dashed')
plt.legend(loc=2)
# plt.savefig('outplots/chaufray_cur2.pdf', facecolor='white', edgecolor='none')
# plt.savefig('chaufray_cur2.png', dpi=400, bbox_inches='tight', facecolor='white')
```

```python
e1 = h2_vel_min[1][1:33] - h2_vel[1,:][1:33]
e2 = h3_vel_min[1][0:32] - h3_vel[1,:][0:32]
e3 = h2_vel[1,:][1:33] - h2_vel_max[1][1:33]
e4 = h3_vel[1,:][0:32] - h3_vel_max[1][0:32]
```

```python
emin = np.sqrt((e1)**2 + (e2)**2)
emin
```

```python
emax = np.sqrt((e3)**2 + (e4)**2)
emax
```

```python
vcurmin = emin + h3_vel[1,0:32]-h2_vel[1,1:33]
vcurmax = h3_vel[1,0:32]-h2_vel[1,1:33] - emax
```

```python
rmax = vcurmax + 0
rmax[rmax < 0] = 0
rmax
```

```python
plt.figure(figsize=(8,6))
ax1 = plt.axes()
ax1.set_ylim(-6, 6)
ax1.set_xlabel('Spatial Axis \nRow Position (0.353$^{\prime\prime}$ per pixel)', fontsize=14)
ax1.set_ylabel('V$_{\mathrm{PRF}}$ (km/s)', fontsize=14)
plt.title('CML = 179', fontsize=14)
plt.plot(h3_vel[0,0:32],h3_int[1,0:32]*1.6+3,color='black',linestyle='dotted',lw=2)
plt.plot(h3_vel[0,0:32],h3_vel[1,0:32]-h2_vel[1,1:33],color='green', label='H$_3^{+}$-H$_2$')
plt.fill_between(h3_vel[0,0:32], vcurmax, vcurmin, color='green', alpha=0.1)
# plt.plot(h3_vel[0,0:32], rmax, color='red', label='H$_3^{+}$-H$_2$ (min)')
plt.plot(rot[0,1:33],rot[1,1:33]-rot[1,1:33],color='black',linestyle='dashed', dashes=(30, 10), lw=0.6)
plt.legend(loc=2)

plt.savefig('chaufray_cur2.png', dpi=400, bbox_inches='tight', facecolor='white')
```
