```python
def read_file(path):
    
    data_list = []
    
    data = sorted(glob.glob(path))
    
    for i in range(len(data)):
        hdu = fits.getdata(data[i])
        data_list.append(hdu)
        
    data_array = np.array(data_list)
    
    return data_array

# filepath = spec_dir+'order36/new_frame/'

order36_data = read_file(spec_dir+'order36/new_frame/*.fits')
order32_data = read_file(spec_dir+'order32/new_frame/*.fits')
order37_data = read_file(spec_dir+'order37/new_frame/*.fits')
order35_data = read_file(spec_dir+'order35/new_frame/*.fits')
order34_data = read_file(spec_dir+'order34/new_frame/*.fits')
```

```python
def calc_avg(order, start, stop, init):
    """
    Calculate the average of data frames.
    
    order: echelle order of spectrograph
    start: index of first fits file to process
    stop: index of last fits file to process
    init: index of first fits file of all
    
    return: the average of ranged fits files in the same shape as individual.
    
    An example: a folder contains fits files from order 36 ranging from 22 to 111 and 
        the files to be processed are 32 to 41. In this case, order = order36, start = 32, 
        stop = 41, init = 22. The order is a self defined variable name, 
        typically an array of 2d spectra data, eg. order.shape = (90, 264, 1024).
        The returned result would be a single 2d data frame in a shape of (264, 1024),
        with each pixel an average of the input fits files at the same pixel position.
    """
    avg = np.mean((order[start-init:stop-init+1]), axis=0)
    
    return avg

int0 = 22
flat0 = 32
flat1 = 41
dark0 = 22
dark1 = 31


avg_flat_order32 = calc_avg(order32_data, flat0, flat1, int0)
avg_flat_order34 = calc_avg(order34_data, flat0, flat1, int0)
avg_flat_order35 = calc_avg(order35_data, flat0, flat1, int0)
avg_flat_order36 = calc_avg(order36_data, flat0, flat1, int0)
avg_flat_order37 = calc_avg(order37_data, flat0, flat1, int0)

avg_dark_order32 = calc_avg(order32_data, dark0, dark1, int0)
avg_dark_order34 = calc_avg(order34_data, dark0, dark1, int0)
avg_dark_order35 = calc_avg(order35_data, dark0, dark1, int0)
avg_dark_order36 = calc_avg(order36_data, dark0, dark1, int0)
avg_dark_order37 = calc_avg(order37_data, dark0, dark1, int0)
```

```python
def fits_header_utc_to_second(path, fn):

    data = sorted(glob.glob(path))
    
    hdu_header = fits.open(data[fn-1], ignore_missing_end=True)[0].header
    
    h = float(hdu_header['UTC'].split(':')[0])
    m = float(hdu_header['UTC'].split(':')[1])
    s = float(hdu_header['UTC'].split(':')[2])
    
    seconds = h * 3600 + m * 60 + s
    
    return seconds

fits_header_utc_to_second(spec_dir+'jun02s*', 61)
```
